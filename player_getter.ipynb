# Creating a Player Database for Players who Participated in the 2018 World Cup

the following code sets up a player database that is used in the pass iq framework

```python
import numpy as np
import pandas as pd
import json
import requests
```

```python
tournament = requests.get('https://raw.githubusercontent.com/statsbomb/open-data/master/data/matches/43.json').json()
```

```python
match_data_url = "https://raw.githubusercontent.com/statsbomb/open-data/master/data/lineups/{}.json"

players = {}

match_ids = []
for g in tournament:
    m_id = g['match_id']
    match_ids.append(m_id)

for id in match_ids:
    all_events = requests.get(match_data_url.format(id)).json()

    for teams in all_events:
        for player in teams['lineup']:
            if player['player_id'] not in players:
                id = player['player_id']
                players[id] = {}
                players[id]['Name'] = player['player_name']
                players[id]['Games'] = {}
                players[id]['Team'] = teams['team_name']
            else:
                pass

        
```
